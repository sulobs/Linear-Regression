```python
# plot best fit line equation 

import numpy as np
import matplotlib.pyplot as plt

x=np.array([8, 9.5, 10, 6, 7, 4])
y=np.array([12, 138, 147, 88, 108, 62])

a, b = np.polyfit(x, y, 1)
plt.xlabel(" X -----> Carbohydrates", color='purple')
plt.ylabel(" Y -----> Calories", color='purple')
plt.scatter(x, y, color='red')
plt.plot(x, a*x+b, color='blue', linestyle='solid', linewidth=2)
plt.text(6, 130, 'y = ' + '{:.2f}'.format(b) + ' + {:.2f}'.format(a) + 'x', size=15)
```
